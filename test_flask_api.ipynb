```python
import requests
import pprint as pprint
# 模擬前端送出的request，用來測試flask api是否有正常運行
url = 'http://192.168.31.80:5000/get-data' #若並非在localhost上運行(可能你的flask開在container或虛擬機上) 請打那台電腦的ip
data = {"name": "ChatGPT", "age": 3}

response = requests.post(url, json=data)
pprint.pprint(response.json())
# 你應該要看到你在flask_titanic.py裡設定的回傳資料
```
